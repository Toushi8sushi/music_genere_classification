```python
from scipy.io import wavfile
import numpy as np
import math as m
import matplotlib.pyplot as plt
```

```python
# Read audio file
#Fs is sampling frequency
#Ts is sampling period 
#N length of data 
Fs, data = wavfile.read('handel.wav')
Ts=1/Fs
N=len(data)


# If stereo, take one channel
if len(data.shape) > 1:
    data = data[:, 0]


# Normalize
data = data.astype(np.float64)
data = data / np.max(np.abs(data))
print(data.shape)
print(Fs)
```

```python
# Make time array
t = np.arange(len(data)) / Fs
print(f'audio length is {t[-1].round(2)} seconds')
```

```python
freq_w_component=np.fft.fft(data)
freq_w_component=np.fft.fftshift(freq_w_component)

abs_freq=abs(freq_w_component)
freq_w=np.linspace(-Fs/2,Fs/2,len(freq_w_component))
freq_w
```

```python
plt.figure()
plt.plot(freq_w,abs_freq)
plt.show()
```

```python
# Load your audio file
y, sr = librosa.load('handel.wav')  # Replace with your file path

# Compute Mel spectrogram
S = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=128, fmax=8000)

# Convert to decibels (log scale)
S_dB = librosa.power_to_db(S, ref=np.max)

# Plot Mel spectrogram
plt.figure(figsize=(10, 4))
librosa.display.specshow(S_dB, x_axis='time', y_axis='mel', sr=sr, fmax=8000)
plt.colorbar(format='%+2.0f dB')
plt.title('Mel-frequency spectrogram')
plt.tight_layout()
plt.show()
```
